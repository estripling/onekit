# `mathkit`

```python
import math

import matplotlib.pyplot as plt

import onekit.mathkit as mk
import onekit.vizkit as vk
```

## `digitscale`

```python
def plot_digitscale(func, bounds, xticks, yticks) -> None:
    _, ax = plt.subplots(1, 1, figsize=(16, 2))
    _, ax, _ = vk.FunctionPlotter(
        func,
        bounds,
        n_xvalues=1001,
        kws_plot={"zorder": 2},
    ).plot(ax=ax)

    # set ticks
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)
    ax.tick_params(axis="x", labelsize=6)
    ax.tick_params(axis="y", labelsize=6)

    # set the facecolor of the axes
    ax.set_facecolor("#EBEBEB")

    # remove border around plot
    [ax.spines[side].set_visible(False) for side in ax.spines]

    # style grid
    ax.grid(which="major", color="white", linewidth=1.2, zorder=0)
    ax.grid(which="minor", color="white", linewidth=0.6, zorder=0)

    ax.set_xlabel(r"$ x $")
    ax.set_ylabel(r"$ f(x) $", rotation=0, labelpad=15)

    plt.show()
```

```python
{x: mk.digitscale(x) for x in [0, 0.1, 0.9, 1, 1.1, 2, 9, 10, 11, 99, 100]}
```

```python
plot_digitscale(
    func=mk.digitscale,
    bounds=[(-0.1, 1)],
    xticks=[-0.1, 0, 0.1, 1],
    yticks=[0, 1],
)
```

```python
plot_digitscale(
    func=mk.digitscale,
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[0, 1, 2],
)
```

```python
plot_digitscale(
    func=mk.digitscale,
    bounds=[(-100, 100)],
    xticks=[-100, -10, 0, 10, 100],
    yticks=[0, 1, 2, 3],
)
```

`digitscale` variants

```python
plot_digitscale(
    func=lambda x: mk.digitscale(x) if x >= 0.0 else 0.0,
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[0, 1, 2],
)
```

```python
plot_digitscale(
    func=lambda x: mk.sign(x) * mk.digitscale(x),
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[-2, -1, 0, 1, 2],
)
```

```python
plot_digitscale(
    func=lambda x: math.floor(mk.digitscale(x)),
    bounds=[(-100, 100)],
    xticks=[-100, -10, 0, 10, 100],
    yticks=[0, 1, 2, 3],
)
```
