# `mathkit`

```python
import math
import textwrap

import matplotlib.pyplot as plt

from onekit import mathkit as mk
from onekit import pythonkit as pk
from onekit import vizkit as vk
```

## `collatz`

```python
def print_collatz(seq: tuple[int]) -> None:
    print(pk.headline(f"n={pk.num_to_str(n)} ({pk.num_to_str(len(seq))} terms)"))
    print(textwrap.fill(pk.concat_strings("  ", map(pk.num_to_str, seq)), width=88))
```

```python
numbers = [27, 54, 97, 6_171, 837_799]

for n in numbers:
    seq = tuple(mk.collatz(n))
    print_collatz(seq)
    plt.plot(seq)
    plt.show()
```

## `digitscale`

```python
def plot_digitscale(func, bounds, xticks, yticks, drawstyle="default") -> None:
    _, ax = plt.subplots(1, 1, figsize=(16, 2))
    _, ax, _ = vk.FunctionPlotter(
        func,
        bounds,
        n_xvalues=1001,
        kws_plot={"drawstyle": drawstyle, "zorder": 2},
    ).plot(ax=ax)

    # set ticks
    ax.set_xticks(xticks)
    ax.set_yticks(yticks)
    ax.tick_params(axis="x", labelsize=6)
    ax.tick_params(axis="y", labelsize=6)

    # set the facecolor of the axes
    ax.set_facecolor("#EBEBEB")

    # remove border around plot
    [ax.spines[side].set_visible(False) for side in ax.spines]

    # style grid
    ax.grid(which="major", color="white", linewidth=1.2, zorder=0)
    ax.grid(which="minor", color="white", linewidth=0.6, zorder=0)

    ax.set_xlabel(r"$ x $")
    ax.set_ylabel(r"$ f(x) $", rotation=0, labelpad=15)

    plt.show()
```

```python
{x: mk.digitscale(x) for x in [0, 0.1, 0.9, 1, 1.1, 2, 9, 10, 11, 99, 100]}
```

```python
plot_digitscale(
    func=mk.digitscale,
    bounds=[(-0.1, 1)],
    xticks=[-0.1, 0, 0.1, 1],
    yticks=[0, 1],
)
```

```python
plot_digitscale(
    func=mk.digitscale,
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[0, 1, 2],
)
```

```python
plot_digitscale(
    func=mk.digitscale,
    bounds=[(-100, 100)],
    xticks=[-100, -10, 0, 10, 100],
    yticks=[0, 1, 2, 3],
)
```

`digitscale` variants

```python
plot_digitscale(
    func=lambda x: mk.digitscale(x) if x >= 0.0 else 0.0,
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[0, 1, 2],
)
```

```python
plot_digitscale(
    func=lambda x: mk.sign(x) * mk.digitscale(x),
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[-2, -1, 0, 1, 2],
)
```

```python
plot_digitscale(
    func=lambda x: mk.digitscale(x, kind="int"),
    bounds=[(-100, 100)],
    xticks=[-100, -10, 0, 10, 100],
    yticks=[0, 1, 2, 3],
    drawstyle="steps-mid",
)
```

```python
{
    x: mk.digitscale(x, kind="linear")
    for x in [0, 0.1, 0.9, 1, 1.1, 2, 9, 10, 11, 99, 100]
}
```

```python
plot_digitscale(
    func=lambda x: mk.digitscale(x, kind="linear"),
    bounds=[(-0.1, 1)],
    xticks=[-0.1, 0, 0.1, 1],
    yticks=[0, 1],
)
```

```python
plot_digitscale(
    func=lambda x: mk.digitscale(x, kind="linear"),
    bounds=[(-10, 10)],
    xticks=[-10, -1, 0, 1, 10],
    yticks=[0, 1, 2],
)
```

```python
plot_digitscale(
    func=lambda x: mk.digitscale(x, kind="linear"),
    bounds=[(-100, 100)],
    xticks=[-100, -10, 0, 10, 100],
    yticks=[0, 1, 2, 3],
)
```
